## 1. (Price) Is there a significant difference in the average price of products between categories that have high ratings vs low ratings?

```python

high_rating_threshold = df['Rating'].quantile(0.75)

high_rating_price = df[df['Rating'] >= high_rating_threshold].groupby('Category')['Price'].mean()
low_rating_price = df[df['Rating'] < high_rating_threshold].groupby('Category')['Price'].mean()


price_diff = (high_rating_price - low_rating_price).reset_index()
price_diff.columns = ['Category', 'Price Difference']

# 1. Filter only significant differences (> 100)
significant_diff = price_diff[price_diff['Price Difference'].abs() > 100]

# 2. Identify categories where higher-rated products are more expensive
higher_price_high_rating = price_diff[price_diff['Price Difference'] > 0]

# 3. Identify categories where higher-1rated products are cheaper
cheaper_price_high_rating = price_diff[price_diff['Price Difference'] < 0]



print("\n‚úÖ Categories with Significant Price Difference (Above ‚Çπ100):")
print(significant_diff)

print("\nüìà Categories where High Rating Products are More Expensive:")
print(higher_price_high_rating)

print("\nüìâ Categories where High Rating Products are Cheaper:")
print(cheaper_price_high_rating)






```

```python
from lets_plot import *
LetsPlot.setup_html()

# Basic bar chart for price difference
ggplot(price_diff.dropna(), aes(x='Category', y='Price Difference')) + \
    geom_bar(stat='identity', fill='blue') + \
    ggtitle('Price Difference Between High and Low Rating Products by Category') + \
    coord_flip()
```

## 2 Question: Do higher-priced products tend to have higher offers?

```python

df['Offer'] = df['Offer'].str.replace('% off', '').astype(float)

high_price_threshold = df['Price'].quantile(0.75)

high_price_offer = df.loc[df['Price'] >= high_price_threshold, 'Offer'].mean()
low_price_offer = df.loc[df['Price'] < high_price_threshold, 'Offer'].mean()

# 1. Check if high-price products have significantly higher offers
if high_price_offer > low_price_offer:
    print(f"High-price products have higher offers: {high_price_offer:.2f}% vs {low_price_offer:.2f}%")
else:
    print(f"Low-price products have higher offers: {high_price_offer:.2f}% vs {low_price_offer:.2f}%")


# 2. Calculate the absolute difference
offer_difference = high_price_offer - low_price_offer
print(f"\nüìä Offer Difference: {offer_difference:.2f}%")

# 3. Display categories where high-price products have better offers
high_price_offer_by_category = df[df['Price'] >= high_price_threshold].groupby('Category')['Offer'].mean()
low_price_offer_by_category = df[df['Price'] < high_price_threshold].groupby('Category')['Offer'].mean()

offer_diff_by_category = (high_price_offer_by_category - low_price_offer_by_category).reset_index()
offer_diff_by_category.columns = ['Category', 'Offer Difference']

# Filter significant differences (>5%)
significant_offer_diff = offer_diff_by_category[offer_diff_by_category['Offer Difference'].abs() > 5]

print("\nüìä Offer Difference by Category:")
print(offer_diff_by_category.sort_values(by='Offer Difference', ascending=False))

print("\n‚úÖ Categories with Significant Offer Differences (>5%):")
print(significant_offer_diff)
```

```python
from lets_plot import *
LetsPlot.setup_html()

# Ensure correct data types
offer_diff_by_category['Category'] = offer_diff_by_category['Category'].astype(str)
offer_diff_by_category['Offer Difference'] = pd.to_numeric(offer_diff_by_category['Offer Difference'], errors='coerce')

# Basic bar chart for offer difference
ggplot(offer_diff_by_category.dropna(), aes(x='Category', y='Offer Difference', fill='Offer Difference')) + \
    geom_bar(stat='identity') + \
    ggtitle('Offer Difference by Category (High vs Low Price)') + \
    coord_flip() + \
    scale_fill_gradient(low='lightblue', high='darkblue')
```

## 3 Question: Do products with better offers tend to have higher ratings?

```python

high_offer_threshold = df['Offer'].quantile(0.75)

high_offer_rating = df[df['Offer'] >= high_offer_threshold]['Rating'].mean()
low_offer_rating = df[df['Offer'] < high_offer_threshold]['Rating'].mean()


# 1. Check if higher offers correlate with higher ratings
if high_offer_rating > low_offer_rating + 0.5:
    print("‚úÖ Products with higher offers tend to have significantly higher ratings.")
elif high_offer_rating > low_offer_rating:
    print("‚úÖ Products with higher offers have slightly higher ratings.")
else:
    print("‚ùå Higher offers do not translate to better ratings.")

# 2. Calculate the absolute difference
rating_difference = high_offer_rating - low_offer_rating
print(f"\nüìä Rating Difference: {rating_difference:.2f}")

# 3. Display by category
offer_rating_by_category = df.groupby('Category')[['Offer', 'Rating']].mean()
offer_rating_by_category['Difference'] = offer_rating_by_category['Offer'] - offer_rating_by_category['Rating']

print("\nüìä Average Offer vs. Rating by Category:")
print(offer_rating_by_category.sort_values(by='Difference', ascending=False))
```

```python
from lets_plot import *
LetsPlot.setup_html()

# Ensure data types are correct
offer_rating_by_category['Category'] = offer_rating_by_category.index.astype(str)
offer_rating_by_category = offer_rating_by_category.dropna()

# Bubble chart (circle chart)
ggplot(offer_rating_by_category, aes(x='Offer', y='Rating', size='Difference', color='Difference')) + \
    geom_point(alpha=0.7) + \
    ggtitle('Average Offer vs Rating by Category') + \
    scale_size(range=[3, 15]) + \
    scale_color_gradient(low='lightblue', high='darkblue')
```

## 4.Are products with more reviews priced higher?

```python
# Define high review threshold
high_review_threshold = df['Reviews'].quantile(0.75)

# Separate high and low review products
high_review_price = df[df['Reviews'] >= high_review_threshold]['Price'].mean()
low_review_price = df[df['Reviews'] < high_review_threshold]['Price'].mean()

# ‚úÖ Add Conditions:
if high_review_price > low_review_price + 50:
    print("‚úÖ Higher-reviewed products are priced significantly higher.")
elif high_review_price > low_review_price:
    print("‚úÖ Higher-reviewed products are slightly more expensive.")
else:
    print("‚ùå Review count does not strongly impact price.")

# 2. Calculate the absolute difference
review_price_difference = high_review_price - low_review_price
print(f"\nüìä Price Difference: ‚Çπ{review_price_difference:.2f}")
```

```python
from lets_plot import *
LetsPlot.setup_html()

# Create a DataFrame for plotting
review_price_comparison = pd.DataFrame({
    'Review Type': ['High Reviews', 'Low Reviews'],
    'Average Price': [high_review_price, low_review_price]
})

# Create the bar chart
ggplot(review_price_comparison, aes(x='Review Type', y='Average Price', fill='Review Type')) + \
    geom_bar(stat='identity', width=0.5, show_legend=False) + \
    ggtitle('üí∏ Average Price of High vs Low Reviewed Products') + \
    scale_fill_manual(values=['#ff7f50', '#87ceeb']) + \
    theme_minimal() + \
    theme(axis_text_x=element_text(size=14, color='#4a4a4a'),
          axis_text_y=element_text(size=14, color='#4a4a4a'),
          plot_title=element_text(size=18, face='bold', color='#4a4a4a')) + \
    geom_text(aes(label=review_price_comparison['Average Price'].round(2)), 
              va='bottom', size=12, color='black')
```
