```python
import pandas as pd
import numpy as np


df = df.rename(columns={
    'Product Title': 'Product_Name',
    'Type': 'Material',
    'Product URL': 'Product_Page',
    'Current Price': 'Price',
    'Discount' : 'Offer',
    'Original Price': 'MRP_Rate',
    'Estimated Delivery': 'Estimated_Delivery'
})


df.insert(0, "ID", np.random.randint(1000, 9999, size=len(df)))

print(df.info())
df.head()
```

```python
import numpy as np

def clean_price(price): 
    if not isinstance(price, str) or price.strip() == "":
        return 0  
    price = price.replace('₹', '').replace(',', '').strip()

    if '-' in price:
        price = price.split('-')[0].strip()  

    try:
        return int(price) 
    except ValueError:
        return 0  


df["Price"] = df["Price"].apply(clean_price)
df["MRP_Rate"] = df["MRP_Rate"].apply(clean_price)
df["Price"] = df["Price"].fillna(0).astype("int32")
df["MRP_Rate"] = df["MRP_Rate"].fillna(0).astype("int32")




print(df.info())  # Check data types
print(df.head())  # Print first few rows to verify





```

```python
import re 

def extract_discount(offer): 
    if not isinstance(offer, str) or offer.strip() == "":
        return 0 
    
    match = re.search(r'(\d+)%', offer) 
    if match:
        return int(match.group(1)) 
    
    return 0


df["Offer"] = df["Offer"].apply(extract_discount)

print(df.info())  
df.head()
```

```python
from datetime import datetime, timedelta

def clean_estimated_delivery(date_str): 
    if not isinstance(date_str, str) or date_str.strip() == "":
        return "No Estimate"

    date_str = date_str.strip()

    if "Delivery Today" in date_str or "Today" in date_str:
        return datetime.today().strftime("%Y-%m-%d")

    if "Delivery Tomorrow" in date_str or "Tomorrow" in date_str:
        return (datetime.today() + timedelta(days=1)).strftime("%Y-%m-%d")
    
    try:
        date_str = date_str.replace("st", "").replace("nd", "").replace("rd", "").replace("th", "") 
        parsed_date = datetime.strptime(date_str, "%a, %d %b")  
        parsed_date = parsed_date.replace(year=datetime.today().year) 
        return parsed_date.strftime("%Y-%m-%d")  
    except ValueError:
        return date_str  

df["Estimated_Delivery"] = df["Estimated_Delivery"].apply(clean_estimated_delivery)

print(df.info())
df.head()
```

```python
import numpy as np

def clean_rating(rating): 
    if not isinstance(rating, str) or rating.strip() == "":
        return 0
    
    try:
        return int(rating.strip())  
    except ValueError:
        return 0 


df["Rating"] = df["Rating"].apply(clean_rating)

print(df.info())
df.head()
```

```python
import numpy as np

def clean_rating(rating): 
    if isinstance(rating, (int)): 
        return rating
    
    if not isinstance(rating, str) or rating.strip() == "": 
        return 0  
    
    try:
        return int(rating.strip()) 
    except ValueError:
        return 0


df["Rating"] = df["Rating"].apply(clean_rating)

print(df.info())
df.tail()
```

```python

def clean_reviews(reviews): 
    if isinstance(reviews, (int)): 
        return reviews
    
    if not isinstance(reviews, str) or reviews.strip() == "": 
        return 0  
    
   
    reviews = ''.join(filter(str.isdigit, reviews))  

    try:
        return int(reviews) if reviews else 0  
    except ValueError:
        return 0 


df["Reviews"] = df["Reviews"].apply(clean_reviews)

print(df.info())
df.tail()
```

```python
import numpy as np

df["Product_Name"] = df["Product_Name"].fillna("Unknown Product").str.strip()
df["Material"] = df["Material"].fillna("Unknown").str.strip().str.title()
df["Product_Page"] = df["Product_Page"].fillna("No Link").str.strip()
df["Category"] = df["Category"].fillna("Uncategorized").str.strip().str.title()



print(df.info())
df.head()
```
