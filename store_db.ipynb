# store database

```python
import pandas as pd
import sqlite3
```

```python
df = pd.read_csv('C:/Project/Data/cleaned_data.csv')
```

```python
conn = sqlite3.connect("vaaree_data.db")  
cursor = conn.cursor()
```

```python
cursor.execute('''
CREATE TABLE IF NOT EXISTS products (
    ID INTEGER PRIMARY KEY,
    Product_Name TEXT,
    Material TEXT,
    Price REAL,
    Product_Page TEXT,
    MRP_Rate REAL,
    Offer TEXT,
    Estimated_Delivery TEXT,
    Category TEXT,
    Rating REAL,
    Reviews INTEGER
)
''')

df.to_sql("products", conn, if_exists="replace", index=False)

conn.commit()
conn.close()
```
